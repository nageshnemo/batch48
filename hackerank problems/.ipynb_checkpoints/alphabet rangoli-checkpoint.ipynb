```python
import string
alpha = string.ascii_lowercase

n = int(input())
L = []
for i in range(n):
    s = "-".join(alpha[i:n])
    L.append((s[::-1]+s[1:]).center(4*n-3, "-"))
print('\n'.join(L[:0:-1]+L))
```

```python
def print_rangoli(size):
    # your code goes here
    k = -1
    char = chr(size+96)
    for i in range(1,size+1,1):
        k+=2
        char = chr(size+96)
        for j in range(1,(size*4-3)+1,1):
            
            if j>=(size*2)-k and j <= (size*2-2)+k:
                if j % 2==0:
                    print("-",end ="")
                else:
                    print(char,end="")
                    if j<size*2-1:
                        char = chr(ord(char)-1)
                    else:
                        char = chr(ord(char)+1)

            else:
                print("-",end="")

        print()
        
    for i in range(1,size,1):
        k-=2
        char = chr(size+96)
        for j in range(1,(size*4-3)+1,1):
            if j>=(size*2)-k and j <= (size*2-2)+k:
                if j % 2 == 0:
                    print("-",end ="")
                else:
                    print(char,end="")
                    if j<size*2-1:
                        char = chr(ord(char)-1)
                    else:
                        char = chr(ord(char)+1)
                    
            else:
                print("-",end="")

        print()
             

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
